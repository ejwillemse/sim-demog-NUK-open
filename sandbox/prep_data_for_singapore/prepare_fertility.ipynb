```python
def interpolation(midyears, popsize):
    x_news = []
    y_news = []
    for i in range(7):
        tempx = []
        tempy = []
            
        if i-1>=0:
            tempx.append(midyears[i-1])
            tempy.append(popsize[i-1])
        else:
            tempx.append(15)
            tempy.append(0)
        tempx.append(midyears[i])
        tempy.append(popsize[i])
        if i+1<7:
            tempx.append(midyears[i+1])
            tempy.append(popsize[i+1])
        else:
            tempx.append(52)
            tempy.append(0)
        #print(i, tempy)
            
        z = np.polyfit(tempx, tempy, 1)
        f = np.poly1d(z)
        x_new = [ float(x) for x in list(range(int(midyears[i]-2), int(midyears[i]+3))) ]
        y_new = f(x_new)
        y_new = [ y if y>0 else 0 for y in y_new ]
        x_news.extend(x_new)
        y_news.extend(y_new)
    return x_news, y_news

#def check
```

```python
origin_years = ['1990', '2000', '2010']
agyears = [col.replace('Years','').replace(' ','').replace('(PerThousandFemales)','') for col in df1.index]
print(agyears)
midyears = [ i*5.+17. for i in range(7) ]
#midyears_ex = [ i*5.+2. for i in range(-1,21) ]
print(midyears)
headers = ['# Given that a birth has a occurred, probability distribution over the    \n',
'# age of the mother. This is generated from ABS age-specific fertility rates,   \n',
'# essentially rescaling them. Theoretically, this is incorrect, as       \n',
'# age-specific fertility rates are expressed as "# births occurring to women of\n',
'# age x" divided by "number of women of age x" However, given that the age\n',
'# distribution is relatively flat over the relevant age range, rescaled     \n',
'# fertility rates serve as a very close approximation.       \n',
'# ---\n', 
'# it is calculated as multiply the female proportion (from female age distribution) by the fertility rate, \n',
'# and calculate the proportion of the multiplication results to sum to 1. \n',  
'# ---\n', 
'#               \n',
'# columns:              \n',
'# 0=age              \n',
'# 1=probability              \n']

for oy in origin_years:
    outpath = os.path.join(out_dir, 'fertility_age_probs_y{}.dat'.format(oy))
    df_temp = df1[[oy]]
    #print(df_temp.head())
    fer = [ f/1000. for f in df1[oy].tolist() ]
    x_news, y_news = interpolation(midyears, fer)
    
    pop = dics[oy]
    y_props = []
    for age, fer in zip(x_news, y_news):
        prop_age = pop[age]
        prop_fer = prop_age * fer
        y_props.append(prop_fer)
    print(y_news[-1], y_props[-1])
    ysum = sum(y_props)
    y_props_parts = [ yy/ysum for yy in y_props ]
    
    with open(outpath, 'w') as fhand:
        for line in headers:
            fhand.write(line)
        for xx, yy in zip(x_news, y_props_parts):
            fhand.write('{} {}\n'.format(str(int(xx)), str(yy)))
```
