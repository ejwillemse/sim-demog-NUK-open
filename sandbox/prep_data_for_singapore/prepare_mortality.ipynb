## prepare data
use 1990, 2000, 2010 as the three input data year, and ~1995, ~2005, ~2015 as the validate year, respectively

```python
df1.head()
```

```python
def interpolation(midyears, popsize):
    x_news = [0]
    y_news = [popsize[0]]
    for i in range(1,19):
        tempx = []
        tempy = []
        ind = []
        if i==1:
            ind = [1,2]
        elif i==18:
            ind = [17,18]
        else:
            ind = [i-1, i, i+1]
        tempx = [midyears[k] for k in ind]
        tempy = [popsize[k] for k in ind]
            
        z = np.polyfit(tempx, tempy, 1)
        f = np.poly1d(z)
        if i==1:
            x_new = [ float(x) for x in list(range(int(midyears[i]-1.5), int(midyears[i]+2.5))) ]
        elif i<18:
            x_new = [ float(x) for x in list(range(int(midyears[i]-2), int(midyears[i]+3))) ]
        else:
            x_new = [ float(x) for x in list(range(int(midyears[i]-2), int(midyears[i]+14))) ]
        y_new = f(x_new)
        x_news.extend(x_new)
        y_news.extend(y_new)
    miny = abs(min(y_news))
    y_news = [ y+miny for y in y_news ]
    return x_news, y_news

#def check
```

## For MALE 
note that the df2 variable name is reuse

```python
sex = 'Male'

df2 = df1.xs(sex, level='gender', drop_level=True)
df2.head()
```

```python
origin_years = ['1990', '2000', '2010']
agyears = [col.replace('Years','').replace(' ','').replace('&Over','~').replace('Under','~').replace('Year','') for col in df2.index]
print(agyears)
midyears = [0, 2.5]+[ i*5.+2. for i in range(1, 18) ]
midyears_ex = [ i*5.+2. for i in range(-1,21) ]
print(midyears)
print(midyears_ex)

for oy in origin_years:
    df_temp = df2[[oy]]
    #print(df_temp)
    dr = [ f/1000. for f in df_temp[oy].tolist() ]
    fig, ax = plt.subplots()
    ax.plot(midyears, dr)
    x_news, y_news = interpolation(midyears, dr)
    
    ax.scatter(x_news, y_news, marker='.')
    ax.set_xticks(midyears_ex)
    ax.set_xticklabels(midyears_ex, rotation=90)
    ax.set_title(oy)
    ax.axvline(x=0, ls=':', c='grey')
    ax.axvline(x=100, ls=':', c='grey')
    ax.axhline(y=0, ls='--', c='grey')
    #print(x_news)
    #print(x_news)
```

```python
origin_years = ['1990', '2000', '2010']
agyears = [col.replace('Years','').replace(' ','').replace('&Over','~').replace('Under','~').replace('Year','') for col in df2.index]
print(agyears)
midyears = [0, 2.5]+[ i*5.+2. for i in range(1, 18) ]
midyears_ex = [ i*5.+2. for i in range(-1,21) ]
print(midyears)
print(midyears_ex)

headers = ['# death rates (male) - data processed from SingStat, check docs/data_input_sources.md for references\n', 
'# columns:\n', 
'#   0=age\n', 
'#   1=death rate (proportion of individuals dying between age x and age x+1\n']

for oy in origin_years:
    outpath = os.path.join(out_dir, 'death_rates_male_y{}.dat'.format(oy))
    df_temp = df2[[oy]]
    #print(df_temp)
    dr = [ f/1000. for f in df_temp[oy].tolist() ]
    x_news, y_news = interpolation(midyears, dr)
    
    with open(outpath, 'w') as fhand:
        for line in headers:
            fhand.write(line)
        for xx, yy in zip(x_news, y_news):
            fhand.write('{} {}\n'.format(str(int(xx)), str(yy)))
    
```
