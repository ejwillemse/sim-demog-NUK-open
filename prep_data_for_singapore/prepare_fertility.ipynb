## prepare data
use 1990, 2000, 2010 as the three input data year, and ~1995, ~2005, ~2015 as the validate year, respectively

```python
def interpolation(midyears, popsize):
    x_news = []
    y_news = []
    for i in range(7):
        tempx = []
        tempy = []
        if i-1>=0:
            tempx.append(midyears[i-1])
            tempy.append(popsize[i-1])
        else:
            tempx.append(14)
            tempy.append(0)
        tempx.append(midyears[i])
        tempy.append(popsize[i])
        if i+1<=6:
            tempx.append(midyears[i+1])
            tempy.append(popsize[i+1])
        else:
            tempx.append(50)
            tempy.append(0)
            
        z = np.polyfit(tempx, tempy, 1)
        f = np.poly1d(z)
        x_new = [ float(x) for x in list(range(int(midyears[i]-2), int(midyears[i]+3))) ]
        y_new = f(x_new)
        x_news.extend(x_new)
        y_news.extend(y_new)
    return x_news, y_news

#def check
```

```python
origin_years = ['1990', '2000', '2010']
agyears = [col.replace('Years','').replace(' ','').replace('(PerThousandFemales)','') for col in df1.index]
print(agyears)
midyears = [ i*5.+17. for i in range(7) ]
#midyears_ex = [ i*5.+2. for i in range(-1,21) ]
print(midyears)
headers = ['# Age-specific fertility rates - data processed from SingStat, check docs/data_input_sources.md for references\n'
'# columns:\n'
'#   0=age\n'
'#   1=fertility rate (proportion of women giving birth between age x and age x+1\n']

for oy in origin_years:
    outpath = os.path.join(out_dir, 'fertility_rates_y{}.dat'.format(oy))
    df_temp = df1[[oy]]
    #print(df_temp.head())
    fer = [ f/1000. for f in df1[oy].tolist() ]
    x_news, y_news = interpolation(midyears, fer)
    
    with open(outpath, 'w') as fhand:
        for line in headers:
            fhand.write(line)
        for xx, yy in zip(x_news, y_news):
            fhand.write('{} {}\n'.format(str(int(xx)), str(yy)))
```
